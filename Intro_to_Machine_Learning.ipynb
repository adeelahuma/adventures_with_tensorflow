- Training is the process through which the model learns how to make sense of the input data.
- Training an Algorithm involves 4 ingredients: 
    - **Data** ; usually historical data 
    - **Model**; simplest model we can train is linear model . in weather forecast example that would mean to 
        find some coefficients , multiply each variable with them and sum everything to get the output 
    - **Objective Function**; estimates how correct the models output are. the entire ML framework boils down to optimizing this function, if function is measuring the prediction error of the model we want to minimize this error or in other words minimize the objective function 
    - **Optimization Algorithm**; it consists of mechanics through which we vary the parameters of the model to optimize the objective function; weights for inputs are the parameters that will change; for each set of parameters we would calculate the objective function than we would choose the model with the highest predictive power

- ML process is iterative; We feed data into model, compare accuracy through objective function, than we vary model parameters and repeat the operation, when we reach a point after which we can no longer optimize or we don't need to we stop

Model 
- f(x) --> y ; y is a function of 'x'
- we don't know this function and want algorithm to find out on its own. 
- this is done by providing it with many pairs of observations of x and y as possible 
- linear model is important and the basis for more complicated models
- In linear model, $f(x) = w.x + b$ ; x = input, w = weight of input/coefficient , b = bias term /intercept
- The goal of the machine learning algorithm would be to find such values for W and B. So the output of X times w plus B is as close to the observed values as possible.
- Example: let's say we want to predict the price of an apartment based on its size, then 
    - then price of 743 sq feet; 336.1  x 743 + 3237.51 = $ 246,484.79
    - then price of 1000 sq feet; 336.1  x 1000 + 3237.51 = $ 332,862.49
    -  the above equation is weight of size * size + bias = price of house 
- Example: let's extend the example to include distance from the beach, then 
    - y = [x1  x2] * [w1 w2] + b = x1.w1 + x2.w2 + b 
    - negative weight; if apartment is farther from the beach then we want to have negative impact on the price 
- Multiple output models; e.g predict purchase price and rent price, output to the model are unchanged 
    - we can create two linear model
        - price = size * wt. of size + proximity_to_beach * wt. of Proximity + bias
        - rent  = size * wt. of size + proximity_to_beach * wt. of Proximity + bias
    - weights/biases of both models will be different     





- Each model is determined solely by the weights and biases; logic of the model remains the same 


Object Function 
- evaluate how well the model's outputs match the desired correct value
- Objective function are generally of two types: 
    - loss /cost ; the lower the loss function, the higher the level of accuracy; mostly in supervised learning  
    - reward; opposite of loss function, The higher the reward function the higher the level of accuracy of the model usually reward functions
        - mostly used in reinforcement learning where the goal is to maximize a specific result 

- Two types of loss function 
    - L2-norm/ squared loss  ; used in regression
    - cross entropy; classification
- Target : desired value at which we are aiming ; labels we assign to data     
- L2-norm $\sum_i (y_i - t_i)^2$ ; the method used to calculate it is called OLS(Ordinary Least Square) 
    - "norm" comes from the fact that it is the vector norm or euclidean distance of the outputs and the targets 
    - the lower the loss; the lower the error
- Cross entropy = $L(y,t)$ = $- \sum_i t_i ln y_i$
    -  t = [0,1,0] or [0,0,1]
    -  y = [0.4, 0.4, 0.2] or [0.1, 0.2, 0.7]
    - L(y,t)= -0* ln0.4 -1* ln 0.4 -0*ln0.2 = 0.92 OR -0* ln0.1 -0* ln 0.2 -1*ln0.7= 0.36 (lower)
    - the lower the loss, the more accurate the model 
- Any function that holds the basic property: higher for worse results, lower for better results can be a loss function 

- Any function that holds the basic property of being higher for worse results and lower for better results can be a loss function. Multiplying by a constant changes nothing. so both of following are true
    - The cross-entropy loss MULTIPLIED by 10 cannot be used for machine learning
    - The cross-entropy loss divided by 10 cannot be used for machine learning
    

Optimization Algorithm 

- the optimization process happens when the optimization algorithm varies the model parameters until the loss function has been minimized.
- in context of linear model this implies varying w and b 
- most fundamental optimization algorithm is 'gradient descent'
- gradient is the multivariate generalization of the derivative; $f\prime(x) = \nabla Fx(x_1, x_2, ...X_n)$
- learning rate : its the rate at which the machine learning algorithm forgets the old belifs for new ones 
- the first derivative of the function is zero when we have reached the minimum
- Oscillation : a repetitive variation around a central value 
- Generally, we want learning rate to be: 
    - High enough, so we can reach the closet minimum in a rational amount of time
    - Low enough, so we are sure we reach the minimum and don't oscillate around the minimum
- update rule : 
    - $x_{i+1} = x_i - \eta f\prime(x_i)$    
    - $w_{i+1} = w_i - \eta \nabla_w L(y,t)$    
    - $b_{i+1} = b_i - \eta \nabla_b L(y,t)$    
    - Now , 
        - $\nabla_w L(y,t) = \nabla_w \sum_i (y_i -t_i)^2$
- key takeaways
    - using gradient descent we can find the minimum of a function by trial and error method
    - update rule allows us to cherry pick the trials so we could reach the minimum faster. Each consequent trial is better than the previous one.
    - must choose learning rate ( high vs low)
    - we should stop updating once we hve converged ; the difference between i and i+1 is 0
